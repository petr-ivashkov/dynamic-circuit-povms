# Load data

```python
[p_bitflip_range, 
 p_non_qnd_range, 
 distances_crem] = pickle.load(open(path+'data/final_results/crem_simulation/hellinger_distances_crem.p','rb'))
```

```python
[p_bitflip_range, 
 p_non_qnd_range, 
 distances_no_crem] = pickle.load(open(path+'data/final_results/crem_simulation/hellinger_distances_no_crem.p','rb'))
```

```python
# This selects a subset of bitflip errors, namely e = 0.1, 0.2 and 0.3
indices_bitflip = [6,8,10]
```

```python
[p_bitflip_range[i] for i in indices_bitflip]
```

# Plot

```python
linestyles = ['--',':','-']
alphas = [1,1,1]
```

```python
# Create a single subplot
fig, ax = plt.subplots(ncols=1, nrows=1, 
                        figsize=(figure_size_x/2, figure_size_y/2),
                        gridspec_kw={'width_ratios': [1]})

linewidth = 0.8
colors = [red, blue, dark_grey]

for i in range(len(indices_bitflip)):
    id = indices_bitflip[i]
    # Results with CREM
    ax.plot(p_non_qnd_range, 
            distances_crem[id], 
            linestyles[i],
            color=blue,
            linewidth=0.8,
            alpha=alphas[i],
            label='$CREM: Ɛ_{Bitflip}$ = '+f'{p_bitflip_range[id]}')
    # Results without CREM
    ax.plot(p_non_qnd_range, 
            distances_no_crem[id], 
            linestyles[i],  
            color=red,
            alpha=alphas[i],
            label='No CREM: $Ɛ_{Bitflip}$ = '+f'{p_bitflip_range[id]}')

# Axes properties
ax.set_xlabel('$Ɛ_{\mathrm{QND}}$')
ax.set_ylabel('Hellinger distance')

# Adjust linewidth of the box
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(linewidth)


fig.tight_layout()
```
