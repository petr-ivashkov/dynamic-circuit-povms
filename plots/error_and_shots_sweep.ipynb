```python
# imports necessary plotting libraries, color definitions, etc. 
from src.utilities.plotting_import import *
```

# Load data

```python
shots_range, infidelities_shots_pauli = pickle.load(open(path+'data/final_results/error_shots_sweep/shots_sweep_pauli.p', 'rb'))
shots_range, infidelities_shots_sic = pickle.load(open(path+'data/final_results/error_shots_sweep/shots_sweep_sic.p', 'rb'))
```

```python
errors_range, infidelities_errors_pauli, infidelities_errors_pauli_std = pickle.load(open(path+'data/final_results/error_shots_sweep/error_sweep_pauli.p', 'rb'))
errors_range, infidelities_errors_sic, infidelities_errors_sic_std = pickle.load(open(path+'data/final_results/error_shots_sweep/error_sweep_sic.p', 'rb'))
```

# Plot

```python
# Create a subplot with 2 rows and 1 column
fig, axs = plt.subplots(ncols=1, nrows=2, 
                        figsize=(figure_size_x/2, figure_size_y),
                        gridspec_kw={'height_ratios': [1,1]})

# Shots sweep
axs[0].loglog(shots_range, infidelities_shots_pauli, '.--', color=red, label='Pauli bases')
axs[0].loglog(shots_range, infidelities_shots_sic, '.--', color=dark_grey, label='SIC-POVM')
axs[0].set_xlabel('Shots')

# Depolariting error sweep
axs[1].errorbar(errors_range, infidelities_errors_pauli, infidelities_errors_pauli_std, fmt='.--', color=red, label='Pauli bases')
axs[1].errorbar(errors_range, infidelities_errors_sic, infidelities_errors_sic_std, fmt='.--', color=dark_grey, label='SIC-POVM')
axs[1].set_xlabel('$\epsilon_{CNOT}$')

# Axis properties
axs[0].set_ylabel('$1-\mathcal{F}_S$')
axs[1].set_ylabel('$1-\mathcal{F}_S$')

# Legend will be added manually in Inkscape
axs[0].legend(loc='upper right',  framealpha=0.8)
axs[1].legend(loc='upper left',  framealpha=0.8)

fig.tight_layout()
```
